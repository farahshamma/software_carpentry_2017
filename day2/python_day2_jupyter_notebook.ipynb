```python
def fahrenheit_celsius(F):
    # C = (F - 32) * (5 / 9)
    celsius = (F - 32) * (5 / 9)
    print(celsius)
    return celsius
```

```python
def get_urban(df):
    urban = df.loc[df['Population_Density'] >= 1000, :]
    return urban
```

whatever is in under loc is in 3rd brackets, it is a conditional arg in the brackets, in this case how may rows you wanna keep how many columns you wanna keep [row, column]. df means it is going to the pandas library. it is a pandas object so it understands that i want some function from pandas library. df is the data frame it is keeping rows of the column pop density that are eq or greater than 1000

```python
def get_urban(df, threshold):
    urban = df.loc[df['Population_Density'] >= 500, :]
    return urban
```

```python
urban_counties = get_urban(dem_health_data, 500)
```

```python
def split_county_by_pop(df, threshold):
    urban = df.loc[df['Population_Density'] >= threshold, :]
    rural = df.loc[df['Population_Density'] < threshold, :]
    return [urban, rural]
```

```python
urban_rural_counties = split_county_by_pop(dem_health_data, 500)
```
