```python
data_for_al = health_data.loc[health_data['State'] == 'AL',\
                              ['Population_Density','Poverty']]
data_for_al.head()
```

```python
# x is a variable here
lst = [tuple(x) for x in data_for_al.values]
```

```python
lst
```

```python
#creating scatter plot stroke = false is the argument to the command; Al in 2nd line is for the legends in the graph
#the 2nd line defines things for the objects created in the 1st line
scatter_plot = pygal.XY(stroke=False)
scatter_plot.add('AL',lst)
SVG(scatter_plot.render())
```

```python
scatter_plot
```

```python
scatter_plot = pygal.XY(stroke=True)
scatter_plot.add('AL',lst)
SVG(scatter_plot.render())
```

```python
#adding title to graph
scatter_plot = pygal.XY(stroke=True)
scatter_plot.title = 'Population Density vs Poverty Correlation'
scatter_plot.add('AL',lst)
SVG(scatter_plot.render())
```

```python
scatter_plot = pygal.XY(stroke=True, x_title='Population Density',\
                       y_title='Poverty')
scatter_plot.title = 'Population Density vs Poverty Correlation'
scatter_plot.add('AL',lst)
SVG(scatter_plot.render())
```
